```python
from bs4 import BeautifulSoup
import requests
import pymongo
from splinter import Browser
import pandas as pd
import time
```

```python
url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
browser.visit(url)

html = browser.html
soup = BeautifulSoup(html, 'html.parser')

soup
```

```python
movies = soup.find('tbody', class_="lister-list").find_all('td', class_="titleColumn")

movies
```

```python
movie_ratings = soup.find('tbody', class_="lister-list").find_all('td', class_="imdbRating")

ratings = []

for rating in movie_ratings:
    ratings.append(rating.text.strip())
# len(ratings)
```

```python
i = 0
rank = []
titles = []
year = []

for movie in movies:
    i += 1
    print(str(i) + ". " + movie.a.text +" ("+ movie.span.text.replace('(','').replace(')','') + ")")
    rank.append(i)
    titles.append(movie.a.text)
    year.append(movie.span.text.replace('(','').replace(')',''))
```

```python
imdb250_df = pd.DataFrame({"rank": rank, "title": titles, "year": year, "rating": ratings})
imdb250_df

imdb250_df.to_csv('Resources/imdb250.csv')

imdb250_df
```
