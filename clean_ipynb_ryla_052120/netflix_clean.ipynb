```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/netflix_titles_raw.csv"
netflix_df = pd.read_csv(csv_file)
netflix_df.head()
```

```python
netflix_df = netflix_df[["title", "release_year"]]
netflix_df
```

```python
csv_file = "Resources/imdb250.csv"
imdb_df = pd.read_csv(csv_file)
imdb_df
```

```python
imdb_netflix = pd.merge(imdb_df, netflix_df, on="title")
imdb_netflix = imdb_netflix[["rank", "id", "title", "rating", "year"]]
imdb_netflix
```

```python
imdb_netflix.to_csv('Resources/netflix_filtered.csv')
```
