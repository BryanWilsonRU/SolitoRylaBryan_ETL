```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/amazonprime_raw.csv"
prime_df = pd.read_csv(csv_file)
prime_df.head()
```

```python
prime_df = prime_df[["title", "year"]]
prime_df
```

```python
csv_file = "Resources/imdb250.csv"
imdb_df = pd.read_csv(csv_file)
imdb_df
```

```python
imdb_prime = pd.merge(imdb_df, prime_df, on="title")
imdb_prime = imdb_prime[["rank", "id", "title", "rating", "year_x"]]
imdb_prime = imdb_prime.rename(columns={"year_x":"year"})
imdb_prime
```
