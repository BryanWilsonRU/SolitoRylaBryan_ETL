```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/hulu_raw.csv"
hulu_df = pd.read_csv(csv_file)
hulu_df.head()
```

```python
hulu_df = hulu_df[["title", "year"]]
print(len(hulu_df))
hulu_df
```

```python
csv_file = "Resources/imdb250.csv"
imdb_df = pd.read_csv(csv_file)
imdb_df
```

```python
imdb_hulu = pd.merge(imdb_df, hulu_df, on="title")
imdb_hulu = imdb_hulu[["rank", "id", "title", "rating", "year_x"]]
imdb_hulu = imdb_hulu.rename(columns={"year_x":"year"})
imdb_hulu
```
