```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Import netflix raw file
csv_file = "Resources/netflix_titles_raw.csv"
netflix_df = pd.read_csv(csv_file)
```

```python
netflix_df
```

```python
# Import IMDB250 file
csv_file = "Resources/imdb250.csv"
imdb_df = pd.read_csv(csv_file)
imdb_df
```

```python
# Match IMDB and netflix movies
imdb_netflix = pd.merge(imdb_df, netflix_df, on="title")
imdb_netflix = imdb_netflix[["rank", "id", "title", "rating_x", "year"]]
imdb_netflix = imdb_netflix.rename(columns={"rating_x":"rating"})
imdb_netflix["service"] = "netflix"
imdb_netflix = imdb_netflix.drop_duplicates(subset="id", keep="first")
imdb_netflix
```

```python
imdb_netflix.describe()
```

```python
imdb_netflix.to_csv('Resources/netflix_filtered.csv')
```
