```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Import csv raw file
csv_file = "Resources/disney_plus_raw.csv"
disney_df = pd.read_csv(csv_file)
disney_df.head()
```

```python
# Clean df
disney_df = disney_df[["title", "year"]]
print(len(disney_df))
disney_df
```

```python
# Import IMDB250 csv
csv_file = "Resources/imdb250.csv"
imdb_df = pd.read_csv(csv_file)
imdb_df
```

```python
# Match IMDB with Disney file
imdb_disney = pd.merge(imdb_df, disney_df, on="title")
imdb_disney = imdb_disney[["rank", "id", "title", "rating", "year_x"]]
imdb_disney = imdb_disney.rename(columns={"year_x":"year"})

imdb_disney = imdb_disney.drop_duplicates(subset="id", keep="first")
imdb_disney["service"] = "disney"
imdb_disney
```

```python
imdb_disney.describe()
```

```python
rg_csv_file = "Resources/disneyplus-reelgood.csv"
disneyplus_url_df = pd.read_csv(rg_csv_file)
disneyplus_url_df.loc[(disneyplus_url_df.title == "Star Wars: A New Hope"),'title']="Star Wars: Episode IV - A New Hope"
disneyplus_url_df.loc[(disneyplus_url_df.title == 'Star Wars: The Empire Strikes Back'),'title']='Star Wars: Episode V - The Empire Strikes Back'
disneyplus_url_df.loc[(disneyplus_url_df.title == 'Star Wars: Return of the Jedi'),'title']='Star Wars: Episode VI - Return of the Jedi'
disneyplus_newdf = pd.merge(imdb_disney, disneyplus_url_df, on='title', how='outer' )
disneyplus_newdf = disneyplus_newdf[["rank", "id", "title", "rating", "year_x", "service", "disneyplus_url"]]
disneyplus_newdf = disneyplus_newdf.rename(columns={"year_x":"year"})

disneyplus_newdf
```

```python
# Save as csv
disneyplus_newdf.to_csv('Resources/disney_filtered.csv')
```
